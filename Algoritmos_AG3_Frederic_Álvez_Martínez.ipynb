##Programación Dinámica. Viaje por el rio
* **Definición**: Es posible dividir el problema en subproblemas más pequeños, guardando las soluciones para ser utilizadas más adelante.
* **Características** que permiten identificar problemas aplicables:<br>
  -Es posible almacenar soluciones de los subproblemas para ser utilizados más adelante<br>
  -Debe verificar el principio de optimalidad de Bellman: “en una secuencia optima de decisiones, toda sub-secuencia también es óptima” (*)<br>
  -La necesidad de guardar la información acerca de las soluciones parciales unido a la recursividad provoca la necesidad de preocuparnos por la complejidad espacial (cuantos recursos de espacio usaremos)<br>

###Problema<br>
En un río hay **n** embarcaderos y debemos desplazarnos río abajo desde un embarcadero a otro. Cada embarcadero tiene precios diferentes para ir de un embarcadero a otro situado más abajo. Para ir del embarcadero i al j, puede ocurrir que sea más barato hacer un trasbordo por un embarcadero intermedio k. El problema consiste en determinar la combinación más barata.

1.jpg

*Consideramos una tabla TARIFAS(i,j) para almacenar todos los precios que nos ofrecen los embarcaderos.<br>
*Si no es posible ir desde i a j daremos un valor alto para garantizar que ese trayecto no se va a elegir en la ruta óptima(modelado habitual para restricciones)

```
#Viaje por el rio - Programación dinámica
################################################################################

TARIFAS = [
[0,5,4,3,float("inf"),999,999],   #desde nodo 0
[999,0,999,2,3,999,11],  #desde nodo 1
[999,999, 0,1,999,4,10], #desde nodo 2
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf") del modulo math
TARIFAS
```

```
#Calculo de la matriz de PRECIOS y RUTAS
#  PRECIOS - contiene la matriz del mejor precio para ir de un nodo a otro
#  RUTAS   - contiene los nodos intermedios para ir de un nodo a otro
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]  #n x n
  RUTA = [ [""]*N for i in [""]*N]

  #Se recorren todos los nodos con dos bucles(origen - destino)
  #  para ir construyendo la matriz de PRECIOS
  for i in range(N-1):
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
```

```
#Asignacion de tareas - Ramificación y Poda
################################################################################
#    T A R E A
#   A
#   G
#   E
#   N
#   T
#   E

COSTES=[[11,12,18,40],
        [14,15,13,22],
        [11,17,19,23],
        [17,14,20,28]]
```

```
#Calculo del valor de una solucion parcial
def valor(S,COSTES):
  """
  Calcula el valor total de una solución dada.
  Una solución es una tupla que indica qué tarea se asigna a cada agente.
  """
  VALOR = 0
  for i in range(len(S)):
    VALOR += COSTES[S[i]][i]
  return VALOR


valor((0, 1, 2, 3  ),COSTES)
```

```
#Coste inferior para soluciones parciales
#  (1,3,) Se asigna la tarea 1 al agente 0 y la tarea 3 al agente 1

def CI(S,COSTES):
  VALOR = 0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range( len(S), len(COSTES)   ):
    VALOR += min( [ COSTES[j][i] for j in range(len(S), len(COSTES))  ])
  return VALOR

def CS(S,COSTES):
  VALOR = 0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range( len(S), len(COSTES)   ):
    VALOR += max( [ COSTES[j][i] for j in range(len(S), len(COSTES))  ])
  return VALOR

CI((0,1),COSTES)
```

```
#Genera tantos hijos como como posibilidades haya para la siguiente elemento de la tupla
#(0,) -> (0,1), (0,2), (0,3)
def crear_hijos(NODO, N):
  HIJOS = []
  for i in range(N ):
    if i not in NODO:
      HIJOS.append({'s':NODO +(i,)    })
  return HIJOS
```

```
from functools import wraps
import time
def calcular_tiempo(funcion):
    """
    Decorador que mide y retorna el tiempo de ejecución en segundos
    """
    @wraps(funcion)
    def wrapper(*args, **kwargs):
        tiempo_inicio = time.time()
        resultado = funcion(*args, **kwargs)
        tiempo_total = time.time() - tiempo_inicio
        return resultado, tiempo_total
    return wrapper
```

```
@calcular_tiempo
def ramificacion_y_poda(COSTES):
    #Construccion iterativa de soluciones(arbol). En cada etapa asignamos un agente(ramas).
    #Nodos del grafo  { s:(1,2),CI:3,CS:5  }
    #print(COSTES)
    DIMENSION = len(COSTES)
    # Inicializamos MEJOR_SOLUCION como una lista con un diccionario
    solucion_inicial = tuple(i for i in range(len(COSTES)))
    MEJOR_SOLUCION = [{'s': solucion_inicial, 'ci': valor(solucion_inicial, COSTES)}]
    CotaSup = MEJOR_SOLUCION[0]['ci']
    #print("Cota Superior:", CotaSup)
    NODOS = []
    NODOS.append({'s': (), 'ci': CI((), COSTES)})

    iteracion = 0

    while len(NODOS) > 0:
        iteracion += 1

        nodo_prometedor = [min(NODOS, key=lambda x: x['ci'])][0]['s']
        #print("Nodo prometedor:", nodo_prometedor)
        #Ramificacion
        #Se generan los hijos
        HIJOS = [{'s': x['s'], 'ci': CI(x['s'], COSTES)}
                for x in crear_hijos(nodo_prometedor, DIMENSION)]

        # Revisamos la cota superior y nos quedamos con la mejor solución
        NODO_FINAL = [x for x in HIJOS if len(x['s']) == DIMENSION]
        if len(NODO_FINAL) > 0:
            #print("\n**Soluciones:",  [x for x in HIJOS if len(x['s']) == DIMENSION  ] )
            if NODO_FINAL[0]['ci'] < CotaSup:
                CotaSup = NODO_FINAL[0]['ci']
                MEJOR_SOLUCION[0] = NODO_FINAL[0]  # Actualizamos manteniendo el formato

        # Poda
        HIJOS = [x for x in HIJOS if x['ci'] < CotaSup]

        # Añadimos los hijos
        NODOS.extend(HIJOS)

        # Eliminamos el nodo ramificado
        NODOS = [x for x in NODOS if x['s'] != nodo_prometedor]

    print("La solucion final es:" ,MEJOR_SOLUCION , " en " , iteracion , " iteraciones" , " para dimension: " ,DIMENSION  )

    return MEJOR_SOLUCION[0]['s'], MEJOR_SOLUCION[0]['ci']#He añadido esto para la actividad 1


ramificacion_y_poda(COSTES)
```

```
from itertools import permutations
@calcular_tiempo
def fuerza_bruta(COSTES):
    mejor_valor = float('inf')  # Inicializamos con infinito
    mejor_solucion = None

    # Generamos todas las posibles permutaciones
    for s in list(permutations(range(len(COSTES)))):
        valor_tmp = valor(s, COSTES)
        if valor_tmp < mejor_valor:
            mejor_valor = valor_tmp
            mejor_solucion = s

    print("La mejor solución es:", mejor_solucion, "con valor:", mejor_valor)
    return mejor_solucion, mejor_valor
```

```
def generar_matriz_costes(dimension, min_valor=10, max_valor=50):
    return [[random.randint(min_valor, max_valor) for _ in range(dimension)]
            for _ in range(dimension)]
```

```
import numpy as np
def ejecutar_comparación(dimensiones, repeticiones=3):#Funcion que compara ambos algoritmos segun su dimension del problema.
    resultados = {}

    for dim in dimensiones:
        print(f"\nProbando dimensión {dim}...")
        tiempos = []
        tiempos2 = []

        for i in range(repeticiones):
            print(f"  Repetición {i+1}/{repeticiones}")
            matriz = generar_matriz_costes(dim)
            if dim<11:
              (solucion, valor), tiempo = fuerza_bruta(matriz)
              tiempos.append(tiempo)
            else:
              tiempos.append(float('inf'))
            (solucion2, valor2), tiempo2 = ramificacion_y_poda(matriz)
            tiempos2.append(tiempo2)

        resultados[dim] = {
            'tiempo_medio_fb': np.mean(tiempos),
            'tiempo_std_fb': np.std(tiempos),
            'tiempos_fb': tiempos,
            'tiempo_medio_rp': np.mean(tiempos2),
            'tiempo_std_rp': np.std(tiempos2),
            'tiempos_rp': tiempos2
        }


    return resultados
```

```
import matplotlib.pyplot as plt
def graficar_resultados(resultados): # Genera una gráfica de los tiempos de ejecución
    dims = list(resultados.keys())
    tiempos_medios = [resultados[d]['tiempo_medio_fb'] for d in dims]
    std_devs = [resultados[d]['tiempo_std_fb'] for d in dims]
    tiempos_medios2 = [resultados[d]['tiempo_medio_rp'] for d in dims]
    std_devs2 = [resultados[d]['tiempo_std_rp'] for d in dims]

    plt.figure(figsize=(10, 6))
    plt.errorbar(dims, tiempos_medios, yerr=std_devs, fmt='o-', capsize=5,label="fuerza_bruta")
    plt.errorbar(dims, tiempos_medios2, yerr=std_devs2, fmt='o-', capsize=5,label="ramificacion_y_poda")
    plt.yscale('log')  # Escala logarítmica para mejor visualización
    plt.xlabel('Dimensión de la matriz')
    plt.ylabel('Tiempo de ejecución (segundos)')
    plt.title('Escalabilidad de los Algoritmos')
    plt.grid(True)
    plt.legend()
    plt.show()
```

$$f(x)=sin(1/2 * x^2 - 1/4 * y^2 + 3) * cos(2*x + 1 - e^y)$$

```
#Definimos la funcion
f= lambda X: math.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) *math.cos(2*X[0] + 1 - math.exp(X[1]) )

def df(X):    #Gradiente, sin funcion anonima ya que es muy largo.
    X0, X1 = X  # Desempaquetamos X en sus componentes

    # Definimos A y B son las fases del seno y coseno
    A = (1/2) * X0**2 - (1/4) * X1**2 + 3
    B = 2 * X0 + 1 - math.exp(X1)

    # Cálculo de las derivadas parciales
    df_dX0 = math.cos(A) * X0 * math.cos(B) - 2 * math.sin(A) * math.sin(B)
    df_dX1 = - (1/2) * X1 * math.cos(A) * math.cos(B) + math.exp(X1) * math.sin(A) * math.sin(B)

    return (df_dX0, df_dX1)
```
